```python
#load libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
youtube_dataset = pd.read_csv('./youtube_dataset.csv')
youtube_dataset.head()
```

```python
# select top 1000 records
top_thousand_records=youtube_dataset.iloc[:1000,:]
top_thousand_records
```

```python
#Create a function to calculate the distribution of channeltype from the top 1000 records.
def cal_dis_fn(dataframe, channeltype):
    print(dataframe[channeltype].count())
    dataframe[channeltype].value_counts().plot(kind='bar', color='indigo')
cal_dis_fn(top_thousand_records, 'channeltype')
```

```python
#Load only the top 1000 records of the original 4000 into a separate CSV file, or database table
top_thousand_records.to_csv('New_1000_Youtubedata.csv')
```
